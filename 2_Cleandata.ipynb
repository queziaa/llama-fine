```python
WOEKTYPE = 3
Finish = {}
import json
fildname = 'outputnew_3.json_Clear.json_Clear.json_Clear.json_Clear.json_Clear.json_Clear.json'
Train = {}
with open('./DATA/train.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
    for i in data:
        Train[i['id']] = i['output']
out = []
with open(fildname + '_Clear.json', 'w', encoding='utf-8') as outf:
    with open(fildname, 'r', encoding='utf-8') as f:
        line = f.readline()
        for line in f.readlines():
            temp = eval(line.strip())
            if WOEKTYPE == 3:
                if temp['id'] in Finish:
                    continue
            elif WOEKTYPE == 4:
               if str(temp['id']) + temp['Target'] + temp['Argument'] in Finish:
                   continue

            results = temp['results']
            results = results.replace('\n1. 俚语分析','\n**俚语分析**').replace('\n2. 语义分析','\n**语义分析**').replace('\n3. 仇恨目标判断','\n**仇恨目标判断**').replace('\n3. 语句片段提取','\n**语句片段提取**')
            results = results.replace('\n1.俚语分析','\n**俚语分析**').replace('\n2.语义分析','\n**语义分析**').replace('\n3.仇恨目标判断','\n**仇恨目标判断**').replace('\n3.语句片段提取','\n**语句片段提取**')
            results = results.replace('**1. 俚语分析**','**俚语分析**').replace('**2. 语义分析**','**语义分析**').replace('**3. 仇恨目标判断**','**仇恨目标判断**').replace('**3. 语句片段提取**','**语句片段提取**')
            results = results.replace('**1. 俚语分析：**','**俚语分析**').replace('**2. 语义分析：**','**语义分析**').replace('**3. 仇恨目标判断：**','**仇恨目标判断**').replace('**3. 语句片段提取：**','**语句片段提取**')
            results = results.replace('**1.俚语分析：**','**俚语分析**').replace('**2.语义分析：**','**语义分析**').replace('**3.仇恨目标判断：**','**仇恨目标判断**').replace('**3.语句片段提取：**','**语句片段提取**')
            results = results.replace('**俚语分析：**','**俚语分析**').replace('**语义分析：**','**语义分析**').replace('**仇恨目标判断：**','**仇恨目标判断**').replace('**语句片段提取：**','**语句片段提取**')
            results = results.replace('**俚语分析：**','**俚语分析**').replace('**语义分析：**','**语义分析**').replace('**仇恨目标判断：**','**仇恨目标判断**').replace('**语句片段提取：**','**语句片段提取**')
            results = results.replace('**俚语分析**','1.俚语分析：').replace('**语义分析**','2.语义分析：').replace('**仇恨目标判断**','3.仇恨目标判断：').replace('**语句片段提取**','3.语句片段提取：')
            results = results.replace('###', '').replace('   - ', '').replace(' ', '').replace('---', '').replace('1.1.','1.').replace('2.2.','2.').replace('3.3.','3.').replace('：：','：')
            results.replace('1.俚语分析', '1.俚语分析：').replace('2.语义分析', '2.语义分析：').replace('3.仇恨目标判断', '3.仇恨目标判断：').replace('3.语句片段提取', '3.语句片段提取：')
            results = results.replace('*','').replace('-','').replace('结论：','').replace('①','').replace('②','').replace('③','').replace('④','').replace('⑤','').replace('⑥','').replace('⑦','').replace('⑧','').replace('⑨','').replace('⑩','')
            if '★' in results:
                print('starerror:', results)
                continue
            if '（注：' in results:
                # print('results:', results)
                results_h = results.split('（注：')
                results_a = results_h[0]
                results_b = results_h[1]
                index = results_b.find('）')
                results_b_b = results_b[index + 1:]
                results = results_a + results_b_b
                deltext = '（注：' + results_b[:index + 1]
                # print('deltext:', deltext)
                # print('results:', results)

            if '\n1.俚语分析：\n' not in results:
                print('liyuerror:', results)
                continue
            if WOEKTYPE == 3:
                if '3.仇恨目标判断：' not in results:
                    print('stageerror:', results)
                    continue
            if WOEKTYPE == 4:
                if '3.语句片段提取：' not in results:
                    print('stageerror:', results)
                    continue
            results = results.split('\n')
            former = []
            paragraph_1 = []
            paragraph_2 = []
            paragraph_3 = []
            stage = 0
            for i in results:
                if len(i) == 0:
                    continue
                if stage == 0:
                    if i != '1.俚语分析：':
                        former.append(i)
                    else:
                        stage = 1
                elif stage == 1:
                    if i != '2.语义分析：':
                        paragraph_1.append(i)
                    else:
                        stage = 2
                elif stage == 2:
                    if i != '3.仇恨目标判断：' and i != '3.语句片段提取：':
                        paragraph_2.append(i)
                    else:
                        stage = 3
                elif stage == 3:
                    paragraph_3.append(i)
            # print('former:', former)
            # print('paragraph_1:', paragraph_1)
            # print('paragraph_2:', paragraph_2)
            # print('paragraph_3:', paragraph_3)
            # print('results:', results)
            # print('-----------------------------')
            if former == [] or paragraph_1 == [] or paragraph_2 == [] or paragraph_3 == []:
                print('nullerror:', results)
                continue
            if WOEKTYPE == 3:
                Finish[temp['id']] = True
            if WOEKTYPE == 4:
                Finish[str(temp['id']) + temp['Target'] + temp['Argument']] = True
            temp['former'] = former
            temp['paragraph_1'] = paragraph_1
            temp['paragraph_2'] = paragraph_2
            temp['paragraph_3'] = paragraph_3
            temp['output'] = Train[temp['id']]
            outf.write(json.dumps(temp, ensure_ascii=False) + '\n')
```
