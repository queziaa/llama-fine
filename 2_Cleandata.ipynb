```python
WOEKTYPE = 4
Finish = {}
import json
fildname = 'outputnew_4_CC.json_Clear.json_Clear.json'
Train = {}
with open('./DATA/train.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
    for i in data:
        Train[i['id']] = i['output']
out = []
with open(fildname + '_Clear.json', 'w', encoding='utf-8') as outf:
    with open(fildname, 'r', encoding='utf-8') as f:
        line = f.readline()
        for line in f.readlines():
            temp = eval(line.strip())
            if WOEKTYPE == 3:
                if temp['id'] in Finish:
                    continue
            elif WOEKTYPE == 4:
               if str(temp['id']) + temp['Target'] + temp['Argument'] in Finish:
                   continue

            results = temp['results']
            results = results.replace('\n1. ä¿šè¯­åˆ†æ','\n**ä¿šè¯­åˆ†æ**').replace('\n2. è¯­ä¹‰åˆ†æ','\n**è¯­ä¹‰åˆ†æ**').replace('\n3. ä»‡æ¨ç›®æ ‡åˆ¤æ–­','\n**ä»‡æ¨ç›®æ ‡åˆ¤æ–­**').replace('\n3. è¯­å¥ç‰‡æ®µæå–','\n**è¯­å¥ç‰‡æ®µæå–**')
            results = results.replace('\n1.ä¿šè¯­åˆ†æ','\n**ä¿šè¯­åˆ†æ**').replace('\n2.è¯­ä¹‰åˆ†æ','\n**è¯­ä¹‰åˆ†æ**').replace('\n3.ä»‡æ¨ç›®æ ‡åˆ¤æ–­','\n**ä»‡æ¨ç›®æ ‡åˆ¤æ–­**').replace('\n3.è¯­å¥ç‰‡æ®µæå–','\n**è¯­å¥ç‰‡æ®µæå–**')
            results = results.replace('**1. ä¿šè¯­åˆ†æ**','**ä¿šè¯­åˆ†æ**').replace('**2. è¯­ä¹‰åˆ†æ**','**è¯­ä¹‰åˆ†æ**').replace('**3. ä»‡æ¨ç›®æ ‡åˆ¤æ–­**','**ä»‡æ¨ç›®æ ‡åˆ¤æ–­**').replace('**3. è¯­å¥ç‰‡æ®µæå–**','**è¯­å¥ç‰‡æ®µæå–**')
            results = results.replace('**1. ä¿šè¯­åˆ†æï¼š**','**ä¿šè¯­åˆ†æ**').replace('**2. è¯­ä¹‰åˆ†æï¼š**','**è¯­ä¹‰åˆ†æ**').replace('**3. ä»‡æ¨ç›®æ ‡åˆ¤æ–­ï¼š**','**ä»‡æ¨ç›®æ ‡åˆ¤æ–­**').replace('**3. è¯­å¥ç‰‡æ®µæå–ï¼š**','**è¯­å¥ç‰‡æ®µæå–**')
            results = results.replace('**1.ä¿šè¯­åˆ†æï¼š**','**ä¿šè¯­åˆ†æ**').replace('**2.è¯­ä¹‰åˆ†æï¼š**','**è¯­ä¹‰åˆ†æ**').replace('**3.ä»‡æ¨ç›®æ ‡åˆ¤æ–­ï¼š**','**ä»‡æ¨ç›®æ ‡åˆ¤æ–­**').replace('**3.è¯­å¥ç‰‡æ®µæå–ï¼š**','**è¯­å¥ç‰‡æ®µæå–**')
            results = results.replace('**ä¿šè¯­åˆ†æï¼š**','**ä¿šè¯­åˆ†æ**').replace('**è¯­ä¹‰åˆ†æï¼š**','**è¯­ä¹‰åˆ†æ**').replace('**ä»‡æ¨ç›®æ ‡åˆ¤æ–­ï¼š**','**ä»‡æ¨ç›®æ ‡åˆ¤æ–­**').replace('**è¯­å¥ç‰‡æ®µæå–ï¼š**','**è¯­å¥ç‰‡æ®µæå–**')
            results = results.replace('**ä¿šè¯­åˆ†æï¼š**','**ä¿šè¯­åˆ†æ**').replace('**è¯­ä¹‰åˆ†æï¼š**','**è¯­ä¹‰åˆ†æ**').replace('**ä»‡æ¨ç›®æ ‡åˆ¤æ–­ï¼š**','**ä»‡æ¨ç›®æ ‡åˆ¤æ–­**').replace('**è¯­å¥ç‰‡æ®µæå–ï¼š**','**è¯­å¥ç‰‡æ®µæå–**')
            results = results.replace('**ä¿šè¯­åˆ†æ**','1.ä¿šè¯­åˆ†æï¼š').replace('**è¯­ä¹‰åˆ†æ**','2.è¯­ä¹‰åˆ†æï¼š').replace('**ä»‡æ¨ç›®æ ‡åˆ¤æ–­**','3.ä»‡æ¨ç›®æ ‡åˆ¤æ–­ï¼š').replace('**è¯­å¥ç‰‡æ®µæå–**','3.è¯­å¥ç‰‡æ®µæå–ï¼š')
            results = results.replace('###', '').replace('   - ', '').replace(' ', '').replace('---', '').replace('1.1.','1.').replace('2.2.','2.').replace('3.3.','3.').replace('ï¼šï¼š','ï¼š')
            results.replace('1.ä¿šè¯­åˆ†æ', '1.ä¿šè¯­åˆ†æï¼š').replace('2.è¯­ä¹‰åˆ†æ', '2.è¯­ä¹‰åˆ†æï¼š').replace('3.ä»‡æ¨ç›®æ ‡åˆ¤æ–­', '3.ä»‡æ¨ç›®æ ‡åˆ¤æ–­ï¼š').replace('3.è¯­å¥ç‰‡æ®µæå–', '3.è¯­å¥ç‰‡æ®µæå–ï¼š')
            results = results.replace('*','').replace('-','').replace('ç»“è®ºï¼š','').replace('â‘ ','').replace('â‘¡','').replace('â‘¢','').replace('â‘£','').replace('â‘¤','').replace('â‘¥','').replace('â‘¦','').replace('â‘§','').replace('â‘¨','').replace('â‘©','')
            if 'â˜…' in results:
                print('starerror:', results)
                continue
            if 'ï¼ˆæ³¨ï¼š' in results:
                # print('results:', results)
                results_h = results.split('ï¼ˆæ³¨ï¼š')
                results_a = results_h[0]
                results_b = results_h[1]
                index = results_b.find('ï¼‰')
                results_b_b = results_b[index + 1:]
                results = results_a + results_b_b
                deltext = 'ï¼ˆæ³¨ï¼š' + results_b[:index + 1]
                # print('deltext:', deltext)
                # print('results:', results)

            if '\n1.ä¿šè¯­åˆ†æï¼š\n' not in results:
                print('liyuerror:', results)
                continue
            if WOEKTYPE == 3:
                if '3.ä»‡æ¨ç›®æ ‡åˆ¤æ–­ï¼š' not in results:
                    print('stageerror:', results)
                    continue
            if WOEKTYPE == 4:
                if '3.è¯­å¥ç‰‡æ®µæå–ï¼š' not in results:
                    print('stageerror:', results)
                    continue
            results = results.split('\n')
            former = []
            paragraph_1 = []
            paragraph_2 = []
            paragraph_3 = []
            stage = 0
            for i in results:
                if len(i) == 0:
                    continue
                if stage == 0:
                    if i != '1.ä¿šè¯­åˆ†æï¼š':
                        former.append(i)
                    else:
                        stage = 1
                elif stage == 1:
                    if i != '2.è¯­ä¹‰åˆ†æï¼š':
                        paragraph_1.append(i)
                    else:
                        stage = 2
                elif stage == 2:
                    if i != '3.ä»‡æ¨ç›®æ ‡åˆ¤æ–­ï¼š' and i != '3.è¯­å¥ç‰‡æ®µæå–ï¼š':
                        paragraph_2.append(i)
                    else:
                        stage = 3
                elif stage == 3:
                    paragraph_3.append(i)
            # print('former:', former)
            # print('paragraph_1:', paragraph_1)
            # print('paragraph_2:', paragraph_2)
            # print('paragraph_3:', paragraph_3)
            # print('results:', results)
            # print('-----------------------------')
            if former == [] or paragraph_1 == [] or paragraph_2 == [] or paragraph_3 == []:
                print('nullerror:', results)
                continue
            if WOEKTYPE == 3:
                Finish[temp['id']] = True
            if WOEKTYPE == 4:
                Finish[str(temp['id']) + temp['Target'] + temp['Argument']] = True
            temp['former'] = former
            temp['paragraph_1'] = paragraph_1
            temp['paragraph_2'] = paragraph_2
            temp['paragraph_3'] = paragraph_3
            temp['output'] = Train[temp['id']]
            outf.write(json.dumps(temp, ensure_ascii=False) + '\n')
```
